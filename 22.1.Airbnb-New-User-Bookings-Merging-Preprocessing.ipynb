**1.IMPORT THƯ VIỆN**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**2.THAO TÁC VỚI DATAFRAME 'SESSIONS'**

```python
sessions = pd.read_csv('/content/drive/MyDrive/Datasets/airbnb-recruiting-new-user-bookings/sessions.csv')
sessions
```

***Thực hiện việc nhóm dữ liệu theo các cột 'user_id, action, action_type, action_detail' trong sessions, sau đó tính toán số lần xuất hiện của mỗi giá trị trong các cột tương ứng:***

```python
session1 = sessions.groupby(["user_id","action"]).agg({"action": "count"})
session2 = sessions.groupby(["user_id","action_type"]).agg({"action_type": "count"})
session3 = sessions.groupby(["user_id","action_detail"]).agg({"action_detail": "count"})
```

```python
#Đổi tên
session1 = session1.rename(columns={'action':'action_count'}).reset_index()
session2 = session2.rename(columns={'action_type':'action_type_count'}).reset_index()
session3 = session3.rename(columns={'action_detail':'action_detail_count'}).reset_index()
```

```python
session1 = session1.pivot_table(index='user_id', columns='action', values='action_count').fillna(0)
session2 = session2.pivot_table(index='user_id', columns='action_type', values='action_type_count').fillna(0)
session3 = session3.pivot_table(index='user_id', columns='action_detail', values='action_detail_count').fillna(0)

#Hợp nhất 3 df lại
session_pivot = session1.merge(session2, on='user_id', how='outer').merge(session3, on='user_id', how='outer')
```

```python
session_pivot
```

**3.THAO TÁC VỚI DATAFRAME 'TRAIN'**

```python
train = pd.read_csv('/content/drive/MyDrive/Datasets/airbnb-recruiting-new-user-bookings/train_users_2.csv')
```

***Tiến hành hợp nhất 2 dataframe 'train' và 'sessions' thành một, đây sẽ là dataframe dùng để training***

```python
df = train.merge(session_pivot, left_on='id', right_on='user_id', how='inner')
df
```

**4.THAO TÁC VỚI DATAFRAME 'TEST'**

```python
test = pd.read_csv('/content/drive/MyDrive/Datasets/airbnb-recruiting-new-user-bookings/test_users.csv')
```

***Tương tự với 'test' và 'sessions' thành một, đây sẽ là dataframe dùng để testing***

```python
df_test = test.merge(session_pivot, left_on='id', right_on='user_id', how='inner')
df_test
```

***1.Xử lý dữ liệu thiếu cho column 'age' cho 2 dataframe:***

*Dễ thấy, cả 2 tập train và test đều có age bất thường, ta sẽ tiến hành: Loại bỏ độ tuổi >100; Những giá trị 'age' nhưng bị nhầm lẫn thành năm sinh sẽ được sửa lại.*

```python
#Hàm loại bỏ dữ liệu vô lý và tính toán dữ liệu ghi nhầm
def age_converter(age):
    if age>1900:
        return (2014-age)
    else:
        return age

df['age'] = df['age'].apply(age_converter)
df_test['age'] = df_test['age'].apply(age_converter)

#Ngoài ra: Tất cả các độ tuổi nhỏ hơn 18 và lớn hơn 100 đều bị chuyển thành NaN
df.loc[df['age']<18, 'age'] = np.nan
df.loc[df['age']>100, 'age'] = np.nan

#Làm tương tự với tập test
df_test.loc[df_test['age']<18, 'age'] = np.nan
df_test.loc[df_test['age']>100, 'age'] = np.nan
```

```python
df["age"] = df["age"].fillna(df.groupby(["signup_method","signup_app","gender"])["age"].transform("mean"))
df_test["age"] = df_test["age"].fillna(df_test.groupby(["signup_method","signup_app","gender"])["age"].transform("mean"))
```

```python
#Làm tròn các giá trị số tuổi
df['age'] = df['age'].astype(int)
```

```python
#tập test vẫn còn giá trị Null, ta sẽ xóa đi 2 giá trị này
df_test = df_test[df_test['age'].notna()]
#Làm tròn các giá trị số tuổi
df_test.loc[:, 'age'] = df_test['age'].astype(int)
```

***2.Xử lý dữ liệu thiếu cho column 'gender' cho 2 dataframe:***

```python
#Đổi tên cho các dữ liệu thiếu cho đồng nhất
df.loc[df['gender'] == 'OTHER', 'gender'] = '-unknown-'
df_test.loc[df_test['gender'] == 'OTHER', 'gender'] = '-unknown-'
```

***3.Xóa dữ liệu Null cho toàn dataframe***

```python
df = df.drop(columns=['date_first_booking'])
df = df.dropna()

df_test = df_test.drop(columns=['date_first_booking'])
df_test = df_test.dropna()
```
